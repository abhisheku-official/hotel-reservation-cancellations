```python
"""
Finding Columns(Numerical Data) which has Maximum values that is more than 2 times it's Mean value,
which provides the infromation that there could be a possible Outlier present in that column.

"""
for column in df:
    if df[column].dtype != 'object':  
        mean_value = df[column].mean()
        max_value = df[column].max()
        if max_value > (mean_value * 2):
            print('Outlier in:', column)
```

```python
# Removing Booking_ID from the dataframe
df.drop('Booking_ID',axis=1,inplace=True)
```

```python
# Finding Unique Data points in each of the Columns
for column in df:
    data=list(df[column])
    unique_data=np.unique(data)
    print(f'COLUMN NAME: {column}\nDATA: {unique_data}\nNo of Unique Data Points: {df[column].nunique()} \n \n')
    
```

```python
# Identifying count of unique categories in each of the Categorical columns.

from collections import Counter

for column in df:
    if  df[column].dtype == "object":  
        count = Counter(df[column])
        print(f' Column: {column}\n Count: {count}\n\n')

        
```

## Exploratory Data Analysis

```python
sns.histplot(data=df, x='no_of_adults', bins=10, discrete=True, hue=df['no_of_adults'].astype(str), multiple='stack')
plt.title('Distribution of no_of_adults')
plt.xlabel('Number of Adults')
plt.ylabel('Count')
plt.show()

"""
# INSIGHT
The majority of room bookings are made for 2 adults, followed by bookings
for 1 adult and 3 adults, suggesting that most travelers are either couples 
or solo travelers, with fewer small group bookings. 
"""
```

```python
sns.histplot(data=df,x='no_of_children', bins=20,discrete=True,hue=df['no_of_children'].astype(str),multiple='stack')
plt.title('Distribution of no_of_children')
plt.xlabel('Number of Children')
plt.show()

"""
#INSIGHT
Most bookings are made without children, suggesting that the majority
of customers are adults traveling alone or in groups.
"""
```

```python
sns.histplot(data=df,x='no_of_weekend_nights', bins=20,discrete=True,hue=df['no_of_weekend_nights'].astype(str),multiple='stack')
plt.title('Distribution of no_of_weekend_nights')
plt.xlabel('Number of Weekend Nights')
plt.show()
"""
#INSIGHT
Most bookings have 0 weekend nights, indicating that customers typically check out before the weekend.
"""
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='no_of_week_nights',bins=30,hue=df['no_of_week_nights'].astype(str))
plt.title('Distribution of no_of_week_nights')
plt.xlabel('Number of Week Nights')
plt.xticks(np.arange(df['no_of_week_nights'].min(), df['no_of_week_nights'].max() + 1))
plt.show()

"""
#INSIGHT
Most bookings include 2 weeknights, followed by 1 and 3 weeknights, indicating that guests 
primarily stay for short weekday trips, possibly for business or midweek travel.
"""
```
