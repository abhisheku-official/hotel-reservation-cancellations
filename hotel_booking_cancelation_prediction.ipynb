## Exploratory Data Analysis

```python
cat_cols = ['type_of_meal_plan', 'required_car_parking_space', 'room_type_reserved',
            'market_segment_type', 'repeated_guest', 'booking_status']
num_cols = ['no_of_adults', 'no_of_children', 'no_of_weekend_nights', 'no_of_week_nights',
           'arrival_year','arrival_month','arrival_date','lead_time','no_of_previous_cancellations', 
          'no_of_previous_bookings_not_canceled','avg_price_per_room','no_of_special_requests']
```

### Univarite Analysis of Numerical Features

```python
def num_plot_dist(df, num_features):
        fig, axes = plt.subplots(len(num_features), 2, figsize=(15, len(num_features) * 5))
        if len(num_features) == 1:
            axes = [axes] 
        for i, column in enumerate(num_features):
            sns.histplot(data=df, x=column, ax=axes[i][0], kde=True)
            axes[i][0].set_title(f'Histogram with KDE for {column}')
            
            sns.boxplot(data=df, x=column, ax=axes[i][1])
            axes[i][1].set_title(f'Box Plot for {column}')
        
        plt.tight_layout()
        plt.show()
```

```python
num_plot_dist(df, num_cols)
```

INSIGHTS

The majority of room bookings are made for 2 adults, followed by bookings for 1 adult and 3 adults, suggesting that most travelers are either couples 
or solo travelers, with fewer small group bookings. 

And most bookings are made without children, suggesting that the majority of customers are adults traveling alone or in groups suggesting a niche market for family-friendly accommodations.

The majority of bookings are for short weekend stays, with a few outliers indicating longer weekend getaways. This suggests a preference for quick weekend escapes and a niche market for extended weekend trips.

Most bookings include 2 weeknights, followed by 1 and 3 weeknights, indicating that guests primarily stay for short weekday trips, possibly for business or midweek travel.

The data indicates a clear seasonal pattern in bookings. There's a significant increase in bookings from July to August, peaking in August. This peak period is then followed by a decline in bookings, suggesting a potential high season for the business during these months.

a majority of rooms being more affordable and a smaller number of high-priced rooms. 

The vast majority of bookings have no special requests, with a few outliers indicating bookings with multiple special requests. This suggests that most customers have simple booking needs, while a small segment requires additional accommodations or preferences.

### Univariate Analysis of Categorical Features

INSIGHT

"Meal Plan 1" is the most popular choice, while "Meal Plan 3" is barely selected. A significant portion of guests did not opt for a meal plan, indicating a preference for outside food or no meal service.

Most guests do not require parking, with only a small fraction requesting it. This suggests that the hotel may not need extensive parking facilities but could target the small segment that does.

"Room Type 1" dominates bookings, while other room types have minimal reservations. The hotel may need to re-evaluate pricing, promotions, or room offerings to increase occupancy for less popular room types.

Online bookings dominate the market, followed by offline bookings, while corporate, complementary, and aviation segments have minimal reservations. To maximize revenue, enhancing online platforms and targeting corporate travelers with tailored packages could be beneficial.


### Bivariate Analysis of Numerical Features

```python
def plot_box_target(df, target, num_features):
    num_plots = len(num_features)
    num_rows = (num_plots + 1) // 2  
    
    fig, axes = plt.subplots(num_rows, 2, figsize=(15, num_rows * 5))
    
    axes = axes.flatten()
    
    for i, column in enumerate(num_features):
        sns.boxplot(x=target, y=column, ax=axes[i], data=df)
        axes[i].set_title(f'{column} vs {target}')
    
    plt.tight_layout()
    plt.show()
```

```python
plot_box_target(df, 'booking_status', num_cols)
```


Number of Adults and Children: These features show similar distributions for

Number of Weekend and Week Nights: Both show similar distributions for canceled

Lead Time: Canceled bookings tend to have longer lead times, implying that longer

cancel again, making past cancellations a strong predictor of future cancellations.

Previous Non-Canceled Bookings: Customers with more non-canceled bookings are less

Number of Special Requests: The number of special requests has little impact on

### Bivariate Analysis of Categorical Features

```python
def plot_cat_features_with_target(df, target, cat_features):
    num_features = len(cat_features)

    num_row = (num_features + 1) // 2

    fig, axes = plt.subplots(num_row, 2, figsize=(12, num_row * 5))
    
    
    axes = axes.flatten()

    for i, feature in enumerate(cat_features):
        sns.countplot(x=feature, hue=target, data=df, palette="Set2", ax=axes[i])
        axes[i].set_title(f'{feature} vs {target}')
        axes[i].set_xlabel(feature)
        axes[i].set_ylabel('Count')
        axes[i].tick_params(axis='x', rotation=90)

    plt.tight_layout()
    plt.show()
```

```python
plot_cat_features_with_target(df, 'booking_status', cat_cols)
```
