```python
"""
Finding Columns(Numerical Data) which has Maximum values that is more than 2 times it's Mean value,
which provides the infromation that there could be a possible Outlier present in that column.

"""
for column in df:
    if df[column].dtype != 'object':  
        mean_value = df[column].mean()
        max_value = df[column].max()
        if max_value > (mean_value * 2):
            print('Outlier in:', column)
```

```python
# Finding Unique Data points in each of the Columns
for column in df:
    data=list(df[column])
    unique_data=np.unique(data)
    print(f'COLUMN NAME: {column}\nDATA: {unique_data}\nNo of Unique Data Points: {df[column].nunique()} \n \n')
    
```

```python
# Identifying count of unique categories in each of the Categorical columns.

from collections import Counter

for column in df:
    if df[column].dtype == 'object':  
        count=Counter(df[column])
        print(f' Column: {column}\n Count: {count}\n\n')
        
```
